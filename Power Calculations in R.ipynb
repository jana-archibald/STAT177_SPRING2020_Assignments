```r
normalplot<-function(m,sd,region=0){
  x<-seq(m-(3.5)*sd,m+(3.5)*sd,length=1000)
  y<-dnorm(x,m,sd)
  plot(x,y,type="l",xlab="",ylim=c(0,max(y)*1.05),xaxs="i",yaxs="i",cex.axis = 2,yaxt='n', ann=FALSE)
  z<-x[x>region[1]]
  z<- z[z<region[2]]
  polygon(c(region[1],z,region[2]),
          c(0,dnorm(z,m,sd),0),col="gray")}
```

# Power Calculations

  ###  The heights of Canandian men are approximately normally distributed with a mean of 174 cm and standard deviation of 12 cm

```r
normalplot(174,12)
```

## $H_0 : \mu_w = 174 $   The mean height of Canadian women is also 174cm

## $H_a : \mu_w < 174 $  The mean height of Canadian women is less than 174 cm

```r
n<- 10
x<-seq(140,190,length=1000)
y<-dnorm(x,174,12/sqrt(n))
plot(x,y,type="l",xlab="",ylim=c(0,max(y)*1.05),xaxs="i",yaxs="i",cex.axis = 2,yaxt='n', ann=FALSE)
lines(x, dnorm(x,164, 12/sqrt(n)), col="red")
```

```r
power.t.test(n=10,
             delta=13,
             sd=12,
             sig.level=0.05,
             power=NULL,
             alternative="one.sided",
             type="one.sample")
```

## What changes with a larger sample

```r
n<- 30
x<-seq(140,190,length=1000)
  y<-dnorm(x,174,12/sqrt(n))
  plot(x,y,type="l",xlab="",ylim=c(0,max(y)*1.05),xaxs="i",yaxs="i",cex.axis = 2,yaxt='n', ann=FALSE)
  lines(x, dnorm(x,164, 12/sqrt(n)), col="red")
```

```r
power.t.test(n=30,
             delta=13,
             sd=12,
             sig.level=0.05,
             power=NULL,
             alternative="one.sided",
             type="one.sample")
```

## What if the difference in height was only 2 cm?

```r
power.t.test(n=30,
             delta=2,
             sd=12,
             sig.level=0.05,
             power=NULL,
             alternative="one.sided",
             type="one.sample")
```

## What if I want to work at $\alpha=0.01$

```r
power.t.test(n=30,
             delta=2,
             sd=12,
             sig.level=0.01,
             power=NULL,
             alternative="one.sided",
             type="one.sample")
```

## How large of a sample do I need...

```r
power.t.test(n=NULL,
             delta=2,
             sd=12,
             sig.level=0.05,
             power=0.99,
             alternative="one.sided",
             type="one.sample")
```
